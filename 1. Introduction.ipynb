1. Supervised Machine Learning:
    - The machine is trained based on experience
    - It learns seeing data with answers, then it tries to predict on unseen data, then based on the error, it tries to learn and imrpovise
    - Two types : 
        - Regression model:
            - The target is a continuous variable:
                - ex. price of a house based on various factors like number of bedrooms, bathrooms, geography etc, 
                - ex. temperature based on historical data and current weather conditions
                - ex. share prices based on historical data and current situation
        - Classification model:
            - The target is a categorical variable:
                - ex. Loan defaulter prediction based on the current credit score, historical data, dti etc
                - ex. Is the tumour malignant or benign based on their shape, size, location
                - ex. based on the x-rays, validate the person has a cancer or not
                - ex. documents classification : pan card, adhar card, voter id, drivers license etc
                - ex. email classification : spam mail, important mail, promotional, social etc
